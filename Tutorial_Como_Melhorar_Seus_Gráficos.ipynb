# Como melhorar a aparência dos seus gráficos usando Python 

Com uso de bibliotecas como *matplotlib* e *seaborn* é simples de gerar gráficos, porém geralmente as configurações padrão nos retornam visualizações feias ou muito simples. Tal fato pode não ser um problema se a imagem gerada é para seus próprios interesses, entretanto se você pretende usar o gráfico para expor algo a outra pessoa é atrativa a ideia de aprender a deixar seus plots com um aspecto melhor.

Essa melhora na aparência do gráfico pode ser bastante importante, pois ajuda a facilitar a quem observa o gráfico perceber quais são os pontos e as informações mais relevantes que a imagem quer passar;

Melhorar um gráfico pode ser feito com algumas mudanças simples, todavia a documentação das biliotecas pode causar dor de cabeça a quem esta dando seus primeiros passos em programação.

Tendo isso em mente vou mostrar nesse post um passo a passo de como transformar um gráfico como esse...

graficoinicial.png

...nesse outro aqui:

graficofinal.png





Vamos começar importando as bibliotecas necessárias.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Agora vamos criar alguns dados imaginários, neste caso criei algo que seria o número de infrações de trânsito durante um dia numa cidade de Minas Gerais.

```
#criação de um vetor para conter os níveis dos delitos
gravidade = ['Leve', 'Média', 'Grave', 'Gravíssima']
#criação de um vetor para conter a quantidade de infrações de cada tipo
ocorrências = [50, 38, 17, 11]
#transformação em dataframe
data = pd.DataFrame(list(zip(gravidade,ocorrências)),columns=['Gravidade','Ocorrências'])
#mostra as primeiras linhas da tabela em questão
data.head()
```

Feito isso passamos a ter uma tabela, ou dataframe, que contém quatro linhas, uma para cada tipo de infração e duas colunas, uma com a gravidade do delito e outra com o número de ocorrências.

A seguir vamos plotar um gráfico padrão usando somente uma linha de código.

```
#plota um gráfico básico
data.plot.bar(x="Gravidade", y="Ocorrências")
```

Mas não está tão caprichado, por isso usaremos agora um gráfico padrão da biblioteca *seaborn*.

```
#plota um gráfico de barras da biblioteca seaborn
sns.barplot(data=data,x='Gravidade', y='Ocorrências')
```

Houve uma melhora considerável, já que foram adicionadas cores diferentes para cada tipo de infração e legendas nos eixos que tornaram o entendimento mais fácil, entretanto ainda há muito à aprimorar.

Começaremos adicionando nossas próprias legendas nos eixos.


```
sns.barplot(data=data, x='Gravidade', y='Ocorrências')
#adiciona a legenda ou rotúlo do eixo x
plt.xlabel('Gravidade da infração') 
#adiciona a legenda ou rotúlo do eixo y
plt.ylabel('Número de infrações')
```

Mas ainda falta uma informação muito importante: o título, pois algo que descreva uma imagem ajuda na recepção da mensagem por parte de quem a vê.


```
sns.barplot(data=data, x='Gravidade', y='Ocorrências')
plt.xlabel('Gravidade da infração')
plt.ylabel('Número de infrações')
#adiciona um título ao gráfico
plt.title('Números de infrações por gravidade')
```

No nosso próximo passo removeremos tanto a legenda do eixo x quanto as bordas do nosso diagrama.

```
sns.barplot(data=data, x='Gravidade', y='Ocorrências')
plt.xlabel('')
plt.ylabel('Número de infrações')
plt.title('Números de infrações por gravidade')
#remove as bordas direita, superior e esquerda do gráfico
sns.despine(left=True);
```

Dando sequência ao trabalho colocaremos linhas horizontas na imagem, para ajudar a visualizar melhor os valores.

```
#torna a imagem mais opaca e adiciona linhas horizontais
sns.set(style='whitegrid')
sns.barplot(data=data, x='Gravidade', y='Ocorrências')
plt.xlabel('')
plt.ylabel('Número de infrações')
plt.title('Números de infrações por gravidade')
sns.despine(left=True);
```

Feita a alteração no corpo da imagem, a próxima etapa é mudar o tamanho da fonte no título e na legenda do eixo vertical.

```
sns.set(style='whitegrid')
sns.barplot(data=data, x='Gravidade', y='Ocorrências')
plt.xlabel('')
#a adição do comando size aumenta a fonte dos textos
plt.ylabel('Número de infrações', size=16)
plt.title('Números de infrações por gravidade', size=16)
sns.despine(left=True);
```

Vamos também mudar o tamanho dos algarismos na eixo vertical.


```
sns.set(style='whitegrid')
sns.barplot(data=data, x='Gravidade', y='Ocorrências')
plt.xlabel('')
plt.ylabel('Número de infrações', size=16)
plt.title('Números de infrações por gravidade', size=16)
#faz mudanças nos rotúlos, pode mudar os valores apresentados,
#nesse caso apenas aumentamos a fonte
plt.xticks(size=14)
plt.yticks(size=14)
sns.despine(left=True);
```

Em seguida vamos alterar as cores das nossas barras, para que possam transmitir a gravidade de cada infração.
Tendo tal ideia em mente usaremos os códigos de cores usando uma ferramenta online, é só buscar por ''colour picker'' no google ou clicar neste [link](https://https://www.google.com.br/search?q=colour+picker&ie=UTF-8&oe=).

```
sns.set(style='whitegrid')
#ao adicionar o comanda palette podemos
#muda a cor de cada barra individualmente
sns.barplot(data=data, x='Gravidade', y='Ocorrências',
            palette=['#0ffc03', '#f0fc03', '#fc7b03', '#eb3434'])
plt.xlabel('')
plt.ylabel('Número de infrações', size=16)
plt.title('Números de infrações por gravidade', size=16)
plt.xticks(size=12)
plt.yticks(size=12)
sns.despine(left=True);
```

Depois a mudança seria feita no tamanho e na resolução da imagem.

```
sns.set(style='whitegrid')
#ajusta o tamanho e a resolução da imagem
plt.figure(figsize=(8,4), dpi=108)
sns.barplot(data=data, x='Gravidade', y='Ocorrências',
            palette=['#0ffc03', '#f0fc03', '#fc7b03', '#eb3434'])
plt.xlabel('')
plt.ylabel('Número de infrações', size=16)
plt.title('Números de infrações por gravidade', size=16)
plt.xticks(size=12)
plt.yticks(size=12)
sns.despine(left=True);
```

Agora adicionaremos uma frase na imagem com intuito de facilitar ao receptor de entender a informação principal que queremos transmitir com o gráfico.

```
sns.set(style='whitegrid')
plt.figure(figsize=(8,4), dpi=108)
sns.barplot(data=data, x='Gravidade', y='Ocorrências',
            palette=['#0ffc03', '#f0fc03', '#fc7b03', '#eb3434'])
plt.xlabel('')
plt.ylabel('Número de infrações', size=16)
plt.title('Números de infrações por gravidade', size=16)
plt.xticks(size=12)
plt.yticks(size=12)
#insere uma linha de texto nas coordenadas x e y específicas 
plt.text(x=1, y=42, s='A maior parte das infrações foi leve ou média', fontsize=10, color='#0f0f0f')
sns.despine(left=True);
```

Nosso gráfico até aqui já está bastante bom, pórem faremos um último ajuste.
Removeremos a legenda do eixo y e colocaremos o número de infrações dentro das barras, este procedimento ajuda a visualizar melhor o número de infrações.
Suponhamos que alguém desejasse saber quantas vezes ocorreram delitos graves? Seria necessário olhar a linha que a barra cria no eixo vertical e em alguns casos tal ação pode ser bastante complicada. 

```
sns.set(style='whitegrid')
plt.figure(figsize=(8,4), dpi=108)
sns.barplot(data=data, x='Gravidade', y='Ocorrências',
            palette=['#0ffc03', '#f0fc03', '#fc7b03', '#eb3434'])
plt.xlabel('')
plt.ylabel('')
plt.title('Números de infrações por gravidade', size=16)
plt.xticks(size=12)
plt.yticks([],[])
plt.text(x=1, y=44, s='A maior parte das infrações foi leve ou média.', fontsize=10, color='#0f0f0f')
sns.despine(left=True);
#análoga à mudança anterior, adiciona texto em locais desejados
plt.text(x=0, y=2, s="50", 
                 color='white', fontsize=18, horizontalalignment='center')
plt.text(x=1, y=2, s="38", 
                 color='white', fontsize=18, horizontalalignment='center')
plt.text(x=2, y=2, s="17", 
                 color='white', fontsize=18, horizontalalignment='center')
plt.text(x=3, y=2, s="11", 
                 color='white', fontsize=18, horizontalalignment='center')
#salva a imagem
plt.savefig('Meugráfico')
```

A adição dos valores nas barras facilita muito o entendimento e compração entre as ocorrências.

E finalmente nosso gráfico está pronto, e convenhamos, muito melhor que o inicial.

É importante que tenhamos em mente que sempre devemos nos atentar ao receptor da nossa mensagem, não será rara a ocasião onde devemos nos comunicar com pessoas que não terão o mesmo conhecimento técnico, por isso é vital que descompliquemos ao máximo a mensagem que desejamos passar.
